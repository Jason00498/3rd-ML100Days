```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '../data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
house_credit_data = app_train[['HOUSETYPE_MODE','AMT_CREDIT']].copy()
house_credit_data['HOUSETYPE_MODE'] = ['None_Type' if pd.isna(x) else x 
                                       for x in house_credit_data['HOUSETYPE_MODE'].values]
house_credit_data['HOUSETYPE_MODE']
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
unique_house_type = house_credit_data['HOUSETYPE_MODE'].unique()

nrows = 2
ncols = 2

plt.figure(figsize=(15,10))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    house_credit_data.loc[house_credit_data['HOUSETYPE_MODE'] == unique_house_type[i], 'AMT_CREDIT'].hist()
    plt.title(str(unique_house_type[i]))
plt.show()    
```
